```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for training and test accuracy across scenarios and models
accuracy_data = {
    "Scenario": [
        "Original", "Original + Tuning", "Resampled", "Resampled + Tuning"
    ] * 3,
    "Model": [
        "Decision Tree", "Decision Tree", "Decision Tree", "Decision Tree",
        "Random Forest", "Random Forest", "Random Forest", "Random Forest",
        "XGBoost", "XGBoost", "XGBoost", "XGBoost"
    ],
    "Training Accuracy (%)": [
        99.81, 99.86, 97.41, 99.75, 100.00, 100.00, 99.79, 99.79, 100.00, 100.00, 99.76, 99.74
    ],
    "Test Accuracy (%)": [
        99.77, 99.83, 97.32, 99.63, 99.85, 99.86, 99.77, 99.77, 99.72, 99.76, 99.72, 99.76
    ]
}

# Create a DataFrame
accuracy_df = pd.DataFrame(accuracy_data)

# Calculate the difference between training and test accuracy
accuracy_df["Difference (%)"] = accuracy_df["Training Accuracy (%)"] - accuracy_df["Test Accuracy (%)"]

# Identify Overfitting or Underfitting
accuracy_df["Status"] = accuracy_df["Difference (%)"].apply(
    lambda x: "Overfitting" if x > 1 else ("Underfitting" if x < -1 else "Good Fit")
)

# Print the results
print(accuracy_df)

# Plot Training vs. Test Accuracy
plt.figure(figsize=(14, 8))
for model in accuracy_df["Model"].unique():
    model_data = accuracy_df[accuracy_df["Model"] == model]
    plt.plot(
        model_data["Scenario"],
        model_data["Training Accuracy (%)"],
        label=f"{model} - Training Accuracy",
        linestyle="--",
        marker="o"
    )
    plt.plot(
        model_data["Scenario"],
        model_data["Test Accuracy (%)"],
        label=f"{model} - Test Accuracy",
        linestyle="-",
        marker="o"
    )

plt.title("Training vs. Test Accuracy Across Scenarios", fontsize=16)
plt.xlabel("Scenario", fontsize=14)
plt.ylabel("Accuracy (%)", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.legend(loc="lower right", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Define parameters for Meezan Bank (PKR Investment)
pkr_investment = 100_000_000  # Initial investment in PKR
pkr_rate = 11.01 / 100  # Annual profit rate
pkr_withdrawal = 100_000 * 12  # Annual withdrawal in PKR
years = 10  # Investment duration

# Define parameters for Germany (EUR Investment)
eur_investment = 300_000  # Initial investment in EUR
eur_rate = 0.28 / 100  # Annual profit rate

# Lists to store yearly balances
pkr_balances = [pkr_investment]
eur_balances = [eur_investment]

# Calculate for Meezan Bank
for _ in range(years):
    # Calculate profit for the year
    profit = pkr_balances[-1] * pkr_rate
    # Subtract withdrawals and reinvest the remaining profit
    balance = pkr_balances[-1] + profit - pkr_withdrawal
    pkr_balances.append(balance)

# Calculate for Germany (EUR Investment)
for _ in range(years):
    # Compound EUR investment annually
    balance = eur_balances[-1] * (1 + eur_rate)
    eur_balances.append(balance)

# Results
final_pkr_balance = pkr_balances[-1]
final_eur_balance = eur_balances[-1]

final_pkr_balance, final_eur_balance
```
